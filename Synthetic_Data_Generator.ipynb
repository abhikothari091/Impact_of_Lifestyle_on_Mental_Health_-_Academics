```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
gender = ['Male','Female']
#regions = ['North America', 'Europe', 'Asia', 'South America', 'Africa', 'Australia', 'Middle East']
universities = ['UMD','NEU','UMass','MIT','Harvard','Stanford','USC','UCSD','TAMU','UTA','NYU','ASU','IUB','PSU','Upenn','CMU','UCI','UCLA']
majors = ['Computer Science', 'Data Science', 'Artificial Intelligence', 'Cyber Security']
degree_levels = ['Undergraduate', 'Graduate']
residential_statuses = ['On Campus', 'Off Campus']
discrimination_status = ['Yes', 'No']
stress_relief_activities = ['Gym', 'Sports', 'Music', 'Writing', 'Travelling','Reading','Running','Cooking','Yoga','Swimming','Gardening']
sleep_ranges = ['2-4 hours','4-6 hours', '6-8 hours', '8-10 hours','10-12 hours']
```

```python
for _, row in original_data.iterrows():
    # Modify every field for synthetic data
    new_row = row.copy()
```

```python
synthetic_data = []
for i in range(num_rows):
    # Modify every field for synthetic data
    new_row = row.copy()
    
    # Randomly generate demographic data
    new_row['gender'] = np.random.choice(gender)        
    new_row['age'] = np.random.randint(18, 30)  # Typical university student age range
    #new_row['Region'] = random.choice(regions)
    new_row['university'] = random.choice(universities)
    new_row['degree_level'] = random.choice(degree_levels)
    new_row['degree_major'] = random.choice(majors)
    new_row['academic_year'] = np.random.choice(['First', 'Second', 'Third', 'Final'])
    new_row['cgpa'] = round(max(0, min(4, np.random.normal(3.5, 0.5))), 2)  # Matching distribution of CGPA
    new_row['residential_status'] = random.choice(residential_statuses)  # New column: Residential Status
    new_row['campus_discrimination'] = random.choice(discrimination_status)  # New column: Campus Discrimination
    new_row['sports_engagement'] = random.choice(['No Sports','1-3 times','4-6 times','7+ times'])  # New column: Sports Engagement
    new_row['average_sleep'] = np.random.choice(sleep_ranges)
    new_row['study_satisfaction'] = np.random.choice([5,4,3,2,1])  # Study Satisfaction
    new_row['academic_workload '] = np.random.choice([3,2,1])
    new_row['academic_pressure'] = np.random.choice([5,4,3,2,1])  # New column: Academic Pressure
    new_row['financial_concerns'] = random.choice([5,4,3,2,1])  # Binary for financial concerns
    new_row['social_relationships'] = random.choice([5,4,3,2,1])  # Binary for social relationships
    new_row['depression'] = random.choice([5,4,3,2,1])  # Random binary for depression
    new_row['anxiety'] = random.choice([5,4,3,2,1])  # Random binary for anxiety
    new_row['isolation'] = random.choice([5,4,3,2,1])  # New column: Isolation
    new_row['future_insecurity'] = random.choice([5,4,3,2,1])  # New column: Future Insecurity
    #new_row['physical_activity'] = np.random.choice([0, 1])  # Random binary values for physical activity
    new_row['stress_relief_activities'] = random.choice(stress_relief_activities)  # New column: Stress Relief Activities

    # Append the modified row to synthetic_data
    synthetic_data.append(new_row)

# Create a DataFrame from synthetic data
synthetic_df = pd.DataFrame(synthetic_data)
```
